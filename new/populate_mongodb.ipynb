```python
import uuid
import random
from datetime import datetime, timedelta, time

def generate_random_data():
    global prompt 
    data = []
    transcripts = []
    start_date = datetime(2022, 1, 1)
    for _ in range(100):  # Generating 10 records as an example
        record = {}
        record['uuid'] = str(uuid.uuid4())
        record['full_name'] = random.choice(italian_names)
        dob = datetime(1990, 1, 1) + timedelta(days=random.randint(0, 365*15))  # Random date within a 15-year range
        record['date_of_birth'] = dob.strftime("%d-%m-%Y")  # Format the DoB as yyyy-mm-dd
        record['change_pw'] = random.choice([True, False])
        record['block_card'] = random.choice([True, False])
        record['opening_hours'] = random.choice([True, False])
        record['appointment'] = random.choice([True, False])
        record['login_issue'] = random.choice([True, False])
        record['card_enable'] = random.choice([True, False])
        record['residence_data'] = random.choice([True, False])
        record['claim_report'] = random.choice([True, False])
        record['card_services'] = random.choice([True, False])
        record['unsolved'] = random.choice([True, False])
        random_days = random.randint(0, (datetime.now() - start_date).days)
        random_time = time(random.randint(9, 20), random.randint(0, 59), random.randint(0, 59))
        record['date_of_call'] = (start_date + timedelta(days=random_days)).strftime("%Y-%m-%d")  # Random date starting from Jan 1, 2022
        record['time_of_call'] = random_time.strftime("%H:%M:%S")  # Random time of the day
        record['call_length_in_sec'] = random.randint(0, 3600)  # Random call length within a 1-hour range
        record['sentiment'] = random.choice(['positive','neutral', 'negative'])

        data.append(record)

        trans = {}
        trans['uuid'] = record['uuid']
        trans['text'] = f"[00:00:00] User:  Buongiorno sono {record['full_name']}, chiamo alle {record['time_of_call']}" + prompt
        transcripts.append(trans)
    return data, transcripts

# Generate random data
random_data, transcripts = generate_random_data()

# Print the generated data
for record in random_data:
    print(record)
```

```python
result = collection_dati.insert_many(random_data)
collection_transcript.insert_many(transcripts)
inserted_ids = result.inserted_ids
# for inserted_id in inserted_ids:
#     print('Inserted document ID:', inserted_id)
```
