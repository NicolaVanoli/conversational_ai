```python
import pandas as pd
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./transcriptions/database.csv')
new_df = df.apply(pd.Series.value_counts).loc[[" 'Yes'"," 'No'"]][['timetables','block_card','change_pw','appointment','login_issue','card_enable','change_residence_data','claim_report','card_services_purchase','unknown']]
new_df.fillna(0, inplace=True)
new_df = new_df.transpose()

fig = new_df.plot.bar(stacked= True,figsize=(13,8),fontsize=16)
fig.legend(fontsize = 16)
plt.show()
```

```python
def my_autopct(pct):
    return ('%1.0f%%' % pct) if pct > 0 else ''

figure, axis = plt.subplots(2,figsize=(10,5),dpi = 140)
ax1 = plt.subplot2grid((1,2),(0,0))
plt.pie(new_df[" 'Yes'"],autopct=my_autopct, pctdistance=.6, textprops={'color':"w"})
plt.legend(new_df.index, loc = 'lower left',bbox_to_anchor=(-.35,-.35))
plt.title('Most Used Service', color = 'white')

ax2 = plt.subplot2grid((1,2), (0, 1))

plt.pie(new_df[" 'No'"],autopct=my_autopct, pctdistance=.6, textprops={'color':"w"})
plt.legend(new_df.index, loc = 'lower left',bbox_to_anchor=(-.35,-.35))
plt.title('Unused Service', color = 'white')

plt.show()
```
